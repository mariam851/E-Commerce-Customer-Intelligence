```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (14, 7)
plt.rcParams['axes.titlesize'] = 12
```

```python
data_path = "../data/E-commerce data.csv"
df = pd.read_csv(data_path, encoding="ISO-8859-1")

df.head()
```

```python
df.info()
```

```python
df.isna().mean().sort_values(ascending=False)
```

```python
# Check invalid transactions
(df["Quantity"] <= 0).sum(), (df["UnitPrice"] <= 0).sum()
```

```python
# Quantity distribution
sns.histplot(df[df['Quantity'].between(0, 100)]['Quantity'], bins=50)
plt.title('Quantity Distribution')
plt.show()
```

```python
# Unit Price distribution
sns.histplot(df[df['UnitPrice'].between(0.1, 50)]['UnitPrice'], bins=50)
plt.title("Unit Price Distribution")
plt.show()
```

```python
# country analysis
top_countries = df['Country'].value_counts().head(10)

sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title("Top 10 Countries by Number of Transactions")
plt.show()
```
